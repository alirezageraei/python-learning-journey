```python
### letter counter ###

def letter_counter(string, char):
    '''
    recives a string and a letter and count
    that letter repeat count in string
    '''

    s = string.lower()
    c = char.lower()
    
    res = 0
    for i in s:
        if i == c:
            res += 1
    
    return res


string = input('write your sentence: ')
char = input('what word should I count? ')
print(letter_counter(string, char))
```

```python
### List comprehension ###

def filter_even(lst):
    '''
    takes a list of numbers and return the even ones
    '''
    return [i for i in lst if i%2==0]

l = [1,2,4,5,6,7,85,]
print(filter_even(l))
```

```python
### variable default ###
''' Mutabe famous trap '''

def add_to_list(item, my_list=None):
    '''
    insets item in the end of recived list
    if list was't exist, makes an empty one and adds item into it 
    '''
    if my_list == None:
        my_list = []
        my_list.append(item)
    else:
        my_list.append(item)
    return my_list


l = list(input('write your items and then press Enter'))
l = [i for i in l if i!=' ']
item = input('write your new item: ')

print(add_to_list(item, l))
```

```python
### Higher-Order Function ###

def apply_to_each(func, lst):
    '''
    takes a Func and a list
    then runs that Func on each item in that list
    '''
    return [func(i) for i in lst]

def square(n):
    return n ** 2

def is_odd(n):
    return n%2==1 

l = [1,2,3,4,5,6,7,8,9]

sq = apply_to_each(square, l)
io = apply_to_each(is_odd, sq)

print(sq)
print([i for i in sq if is_odd(i)])
```

```python
### closure finction ###
### توابع حافظه دار (بسیار جالب) ###
def make_counter():
    '''
    everytime you call this function, counter will increases 1 Unit
    '''
    count = 0
    def counter():
        nonlocal count
        count += 1
        return count
    return counter

counter = make_counter()

print(counter())
print(counter())
print(counter())
```

```python
### smart Functions ###
### کاملا کاربردی و کمی عمیق ###
def smart_average(*args, **kwargs):
    if kwargs.get('weighted'):
        total_weighted = 0
        total_weight = 0
        
        for arg in args:
            if isinstance(arg, (list, tuple)) and len(arg) == 2:
                value, weight = arg
                total_weighted += value * weight
                total_weight += weight
            else:
                print(f"Skipping invalid weighted pair: {arg}")
        
        if total_weight == 0:
            return 0
        return total_weighted / total_weight
    
    lst = []
    for arg in args:
        if isinstance(arg, (int, float)):
            lst.append(arg)
        elif isinstance(arg, (list, tuple)):
            for i in arg:
                if isinstance(i, (int, float)):
                    lst.append(i)
                else:
                    print(f"one of entered values is not in the right type: {i}")
    
    if len(lst) == 0:
        return 0
    return sum(lst) / len(lst)


print(smart_average(1,2,3,4,5,6,7,9,10.5,2))
print(smart_average([1,2,3,4,5,6,7,9,10.5],2))
print(smart_average((1,1000),(3,5),(6,11),(30,2),weighted=True))
```

```python
### dict finder ###

def find_dict(args, name):
    for i in args:
        if isinstance(i, dict) and  i["name"] == name:
                return i
    return None


lst = [
    {'name': 'ali', 'age': 25},
    {'name': 'reza', 'age': 40},
    {'name': 'hossein', 'age': 20},
    {'name': 'sara', 'age': 22},
    {'name': 'zahra', 'age': 30},
    {'name': 'samira', 'age': 19}
    ]
```
