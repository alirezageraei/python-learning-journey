## Infinite Counter Generator

This generator produces an endless sequence of numbers starting from a given value.

### Concepts Practiced
- Basic generator syntax (`yield`)
- Infinite loops with generators
- Lazy evaluation
- Using `next()` manually

### Why It Matters
Infinite generators are useful when working with streams of data that do not have a predefined end.

```python
### infinite counter ###

def counter(start):
    """
    get's a value and start counting from that value to Inf,
    increasing by 1 for each step
    """
    c = start
    while True:
        yield c 
        c += 1

n = counter(56)

for i in range(10):
    print(next(n))
```

## Even Number Generator

This generator yields only even numbers up to a specified limit.

### Concepts Practiced
- Conditional yielding
- Generator termination using `break`
- Filtering generated values

### Why It Matters
Generators can be used to efficiently filter large sequences without storing all values in memory.

```python
### Even nomber generator ###

def even_generator(limit):
    """
    get's a value and start counting from 0 to that value,
    but, it just returns Even ones
    """
    for num in range(0, limit+1, 2):
        yield num

for item in even_generator(10):
    print(item)
```

## Lazy Reader

This generator returns items from a list one at a time instead of returning the entire collection at once.

### Concepts Practiced
- Lazy evaluation
- Iteration using generators
- Memory-efficient data access

### Why It Matters
Lazy loading is useful when working with large datasets where loading everything at once would be inefficient.

```python
### Lazy Reader ###

def lazy_reader(lst):
    """
    reads a list and return it's items,
    one each time
    """
    if isinstance(lst, list):
        for item in lst:
            yield item
    else:
        raise TypeError("input ,must be a list")    
    
l = [10,30,45,60,75]

data = lazy_reader(l)
print(next(data))
# print(next(data))
# print(next(data))
```

## Prime Number Generator

This generator yields prime numbers up to a given limit.

### Concepts Practiced
- Nested loops
- Conditional logic
- Generator-based number filtering
- Basic algorithm optimization

### Why It Matters
Prime number generation is a classic programming exercise that improves problem-solving and algorithmic thinking.

```python
### Prime-numbers generator ###

def prime_generator(limit):
    """
    gets a value and returns the Prime-numbers from 0 to limit value
    """
    if isinstance(limit, int) and limit >= 0:
        for item in range(2, int(limit ** 0.5) + 1): # for optimizing, we just check numbers smaller than our limit's seccond root. reason --> math rules
            is_prime = True
            for num in range(2, item):
                if item % num == 0:
                    is_prime = False
                    break # instant exit of loop: optimizing tip
            if is_prime:
                yield item
    else:
        raise ValueError("Input should be a Positive Integer")
        
primes = prime_generator(50)

print(next(primes))
print(next(primes))
print(next(primes))
```

## File Line Reader

This generator reads a file line by line instead of loading the entire file into memory.

### Concepts Practiced
- File handling
- Generators with external resources
- Memory-efficient file processing
- Exception handling

### Why It Matters
Reading files lazily is essential when working with large files and data-processing pipelines.

```python
### read files line by line ###

def read_lines(filename):
    try:
        with open(filename, 'r') as f:
            for line in f:
                yield line
                """ instead of loop we can use this:
                    "yield from f"
                """
    except FileNotFoundError as e:
        raise

lines = read_lines("grades.csv")

print(next(lines))
print(next(lines))
print(next(lines))
```

## Error Log Filter

This generator extracts only error messages from a collection of log entries.

### Concepts Practiced
- Data filtering
- String processing
- Conditional yielding

### Why It Matters
Log filtering is a common real-world task used for debugging and monitoring applications.

```python
### Error Log Filter ###

def error_logs(logs):
    if isinstance(logs, list):
        for log in logs:
            if "ERROR" in log.upper(): # better to use: if log.upper().startswith("ERROR")
                yield log
    else:
        raise TypeError("error_logs function can only be used for lists")

logs = [
    "INFO: start",
    "ERROR: db failed",
    "INFO: finish",
    "ERROR: connection timeout",
]

for i in error_logs(logs):
    print(i)
```

## Fibonacci Generator

This generator produces Fibonacci sequence values up to a specified limit.

### Concepts Practiced
- State management inside generators
- Multiple variable updates
- Mathematical sequences

### Why It Matters
Fibonacci generators demonstrate how generators can maintain state between iterations.

```python
### Fibonacci Generator ###

def fibonacci(limit):
    """
    Generate Fibonacci numbers smaller than limit.
    """
    first = 0
    seccond = 1

    while True: # condition can be also: first < limit (when we want to stop generating limited)
        yield first
        first, seccond = seccond, first + seccond

fib = fibonacci(50)

for _ in range(10):
    print(next(fib))
```

## Generator Pipeline

This exercise chains multiple generators together to transform data step by step.

### Concepts Practiced
- Generator composition
- Data pipelines
- Lazy transformations

### Why It Matters
Generator pipelines are commonly used in data processing systems to efficiently handle large streams of information.

```python
### Simp1e Generator Pipeline ###

def number_generator(n):
    """
    Generate numbers from 1 to n
    """
    for number in range(1, n + 1):
        yield number

def square_numbers(generator):
    """
    return the square of each number from the input generator.
    """
    for i in generator:
        yield i ** 2

num = number_generator(10)
sq = square_numbers(num)

for i in sq:
    print(i)
```
