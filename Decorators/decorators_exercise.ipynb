## Before & After Decorator

A simple decorator that executes code before and after a function call.

### Concepts Practiced
- Basic decorator structure
- Function wrapping
- Nested functions

### Why It Matters
Many real-world decorators add behavior before and after a function executes, such as logging and authentication.

```python
### Before & After Decorator ###
def decorator(func):

    def wrapper():
        print("Starting Function...")

        func()

        print("Function Finished.")

    return wrapper

def greet():
    print(f"Hello World")
    
say_hello = decorator(greet)
say_hello()
```

```python
### Standard format ###

def my_decorator(func):

    def wrapper(*args, **kwargs):

        print("Before")

        res = func(*args, **kwargs)

        print("Before")

        return res

    return wrapper


@my_decorator
def say_hello(name):
    print(f"Hello {name}")


say_hello("Ali")
```

## Call Counter Decorator

A decorator that keeps track of how many times a function has been executed.

### Concepts Practiced
- Closures
- State preservation
- Nonlocal variables

### Why It Matters
Closures allow decorators to store information between function calls without using global variables.

```python
### Call Counter Decorator ###
""" Stateful decorator:
    A decorator that can memorize information between different executions
"""
def call_counter(func):
    count = 0
    def wrapper(*args, **kwargs):
        nonlocal count
        count += 1
        print(f"Function {func.__name__} called {count} times")
        res = func(*args, **kwargs)
        return res
    return wrapper

@call_counter
def Passcheck():
    print(f"Wrong password")

Passcheck()
Passcheck()
Passcheck()
```

## Uppercase Output Decorator

A decorator that modifies the return value of a function by converting strings to uppercase.

### Concepts Practiced
- Working with return values
- Output transformation
- Wrapper functions

### Why It Matters
Decorators can modify both the behavior and output of existing functions without changing their code.

```python
### Uppercase output Decorator ###

def uppercase(func):

    def wrapper(*args, **kwargs):

        res = func(*args, **kwargs)
        if isinstance(res, str):
            return res.upper()
    return wrapper

@uppercase
def get_name():
    return "alireza"

get_name()
```

## Timer Decorator

A decorator that measures and displays a function's execution time.

### Concepts Practiced
- Performance measurement
- External modules (`time`)
- Function wrapping

### Why It Matters
Execution timing is useful for performance optimization and benchmarking.

```python
### Timer decorator: start to use some decorator applicatins ###

import time

def timer(func):

        def wrapper(*args, **kwargs):
            start = time.time()

            result = func(*args, **kwargs)

            end = time.time()

            run_time = end - start
            print(f"Run time: {run_time :.7f}")
            
            return result
        return wrapper


@timer
def multiplier(a, b, c):
      return a * b * c

mult = multiplier(100, 42, 371)
print(f"Resullt: {mult}")
```

## Logger Decorator

A decorator that logs information about function execution.

### Concepts Practiced
- Function metadata
- Using `__name__`
- Logging function calls

### Why It Matters
Logging helps developers monitor, debug, and understand program behavior.

```python
### Logger Decorator ###

def logger(func):
    def wrapper(*args, **kwargs):
        
        print(f"Running function: {func.__name__}")
        print(f"Arguments: {args}")
        print(f"Keyword Arguments: {kwargs}")
        
        res = func(*args, **kwargs)
        return res
    return wrapper

@logger
def add(a, b):
    return a + b

add(10, 20)
```

## Repeat Decorator

A parameterized decorator that executes a function multiple times.

### Concepts Practiced
- Decorator factories
- Closures
- Multiple nested functions

### Why It Matters
Decorators can accept arguments and dynamically change their behavior.

```python
### repeat decorator ###
### Nested decorators (Decorators with input) ###

def repeat(repeats):
    """ first layer: decorator input value """
    def deforator(func):
        """ second layer: decorator input function """
        def wrapper(*args, **kwargs):
            """ third layer: Wrapper """
            print("Function launched!...")
            
            for _ in range(repeats):
                func(*args, **kwargs)
            
            print("Function ended!")
            
            
        return wrapper
    return deforator

"""if time was in repeat place, it would the time of each repeat,
    But in this place(above repeat) it will take the whole execution time."""
@timer
@repeat(5)
def welcome(name):
    print(f"Welcome {name}")

Ali = welcome("Ali")
Ali
```

## Type Checker Decorator

A decorator that validates input types before executing a function.

### Concepts Practiced
- Input validation
- Type checking
- Defensive programming

### Why It Matters
Input validation helps prevent unexpected errors and improves code reliability.

```python
### Type Checker Decorator ###

def type_check(func):

    def wrapper(*args, **kwargs):
        check = True
        for arg in args:
            if not isinstance(arg , int):
                raise TypeError("All arguments must be integers")
        for value in kwargs.values():
            if not isinstance(value , int):
                raise TypeError("All arguments must be integers")
        if check:
            res = func(*args, **kwargs)
            return res
    return wrapper

@type_check
def divide(a, b):
    return a / b

try:
    print(divide(100, 2))
except ZeroDivisionError:
    print("Division by Zero is not allowed")
```

## Input Validator Decorator

A decorator that checks user input against predefined rules.

### Concepts Practiced
- Data validation
- Exception handling
- Function interception

### Why It Matters
Validation is a common requirement in applications that process user input.

```python
### Input Validator Decorator ###

def validator(func):

    def wrapper(*args, **kwargs):
        for arg in args:
            if len(arg) == 0:
                raise ValueError("Username cannot be empty")
            if len(arg) < 8:
                return f"Username must be at least 8 character,'{arg}' is {len(arg)}"
        for value in kwargs.values():
            if len(value) == 0:
                raise ValueError("Username cannot be empty")
            if len(value) < 8:
                return f"Username must be at least 8 character,'{value}' is {len(value)}"
        
        res = func(*args, **kwargs)
        return res
    return wrapper

@validator
def login(username):
    print("Logged in")

print(login("z"))
```
