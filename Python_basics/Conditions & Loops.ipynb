```python
### Build max and min func ###
def Max(Numbers):
    max = 0
    for i in Numbers:
        if i >= max:
            max = i
    return max

def Min(Numbers):
    min = 0
    for i in Numbers:
        if i >= min:
            min = i
    return min

l = [12, 5, 8, 20, 3, 15, 7]

print(Max(l))
print(Min(l))
```

```python
### nested lists and loops ###

l = [[3, 7, 2], [5, 9, 1], [4, 6, 8]]

Il = len(l)
Rl = len(l[0])

Imax = [max(i) for i in l]
print(Imax)

Sum = [0, 0, 0]

for i in range(Il):
    for j in range(Rl):
        Sum[i] = Sum[i] + l[j][i]

print(Sum)
```

```python
### Set Algebra ###

group_a = ["ali@mail.com", "sara@mail.com", "reza@mail.com"]
group_b = ["sara@mail.com", "mina@mail.com", "ali@mail.com"]


setA = set(group_a)
setB = set(group_b)


common = setA.intersection(setB)
difference = setA.difference(setB)


output = {
    'common': list(common),
    'only_a': list(difference)
}

print(output)
```

```python
### players ranking ###

from random import randint as r

scores = { "player1": [12, 8, 15, 10], "player2": [14, 9, 8, 12], "player3": [10, 11, 13, 9] }

avg_of_scores = [sum(i)/len(i) for i in scores.values()]
best_player = avg_of_scores.index(
    max(avg_of_scores)
)

for i in scores:
    x = r(5, 20)
    scores[i].append(x)
  
avg_of_scores_ = [sum(i)/len(i) for i in scores.values()]
best_player_ = avg_of_scores_.index(
    max(avg_of_scores_)
)

print(f'sum of scores is: {avg_of_scores}')
print(f'the highest average score, belongs to player {best_player+1}\n')
print('*'*50)
print(f'New sum of scores is: {avg_of_scores_}')
print(f'the New highest average score, belongs to player {best_player_+1}')

if best_player != best_player_:
    print("Rank has changed!")
else:
    print("Rank is still the same")
```

```python
### Bubble sort ###

def mini(args):
    minimum = float("inf")
    for i in args:
        if i < minimum:
            minimum = i
    return minimum

    
num_list = [5,4,8,72,15,4,2,41,12,42,34,51,11,20,10,9,44,33,16,101]

num_list_copy = num_list.copy()
sorted_list = []
for i in range(len(num_list_copy)):
    min_num = mini(num_list_copy)
    sorted_list.append(min_num)
    num_list_copy.remove(min_num)

print(num_list) 
print(sorted_list)
```

```python
### Sliding window Algorythm ###
### الگوریتم پنجر لغزنده (کاربردی) ###
    
n = int(input("Enter n: "))

l = [i for i in range(1, n)]

start = 0
end = 0
s = 0

while start < n-1:
    while s < n and end < n-1:
        s += l[end]
        end += 1
    
    if s == n:
        print(f"{n} = {l[start:end]}")
        break
    
    s -= l[start]  # حذف عدد start از مجموع
    start += 1

if s != n:
    print("یافت نشد")
```
