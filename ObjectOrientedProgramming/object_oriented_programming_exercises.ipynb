## 1. Creating a Simple Class

A basic example of defining a class and implementing instance methods.

```python
### making a s1mple class ###

class Dog():
    def __init__(self,name,race):
        self.name = name
        self.race = race

    def bark(self):
        if self.race == "poppy":
            print("Yap!")
        else:
            print("Woof!")
    
    def introduce(self):
        article = "an" if self.race[0].lower() in "aeiou" else "a"
        print(
            f"Hi I am {self.name} and I am {article} {self.race}\n"
            + "_" * 30
        )
german = Dog("kaity", "German-Shepard")
german.bark()
german.introduce()

doberman = Dog("rex", "O-doberman")
doberman.bark()
doberman.introduce()        
```

## 2. Rectangle Class

A simple class that calculates area and perimeter.

```python
### S1mple calculation class ###
class Rectangle():

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return f"the area of this rectangle is {self.width * self.length}"
    
    def perimeter(self):
        return f"the perimeter of this rectangle is {2 * (self.width + self.length)}"
    
rec1 = Rectangle(10,5)
print(rec1.area())
print('-'*40)
print(rec1.perimeter())
    
```

## 3. Bank Account Simulation

Demonstrates validation, deposits, withdrawals, and class attributes.

```python
### Bank Account ###

class BankAccount():

    account_name = "Bank Melli Iran | Short-term loan account"
    def __init__(self, Start_credit):
        if isinstance(Start_credit, (int, float)):
            self.current_balance = Start_credit
        else:
            self.current_balance = 0
        print(f"your account has been made | Welcome to our bank system\naccount info: {self.account_name}\n"+"_"*30)

    def deposit(self, Additional_credit):
        self.current_balance += Additional_credit
        return f"You added {Additional_credit} in your account\nyour current balance is: {self.current_balance}\n"+"_"*30
  
    def withdraw(self, withdraw_credit):
        if self.current_balance >= withdraw_credit:
            self.current_balance -= withdraw_credit
            return f"You picked {withdraw_credit} out of your account\nyour current balance is: {self.current_balance}\n"+"_"*30
        else:
            raise ValueError("You can't withdraw more than your account balance")

my_account = BankAccount(5000)

print(my_account.current_balance)
print(my_account.deposit(2000))
print(my_account.withdraw(800))
print(my_account.withdraw(9000)) # Error
```

## 4. Basic Inheritance

Creating a child class from a parent class and extending functionality.

```python
### s1mple inheritance ###

class Car():

    def __init__(self, company, model, brand):
        self.maker = company
        self.model = model
        self.brand = brand
    
    def display_info(self):
        return f"This car is a {self.brand} from company {self.maker}, the model of thisone is {self.model}\n"+"-"*50


class ElectricCar(Car):

    def __init__(self, company, model, brand, battery_health):
        super().__init__(company, model, brand)
        if isinstance(battery_health, int):
            self.battery = battery_health
        else:
            raise TypeError("***Battery health must be an Integer***")

    def display_info(self):
        baseinfo = super().display_info().strip()
        return f"{baseinfo}\nThisones battery health percentage is {self.battery}%\n" + "-" * 30


car1 = Car("Mercedes", "Mx55", "Maibackh")
print(car1.display_info())

car2 = ElectricCar("Tesla", "ST2000", "CyberTruck", 98)
print(car2.display_info())
```

## 5. Aggregation

Books can exist independently from the library that stores them.

```python
### Aggregation ###
""" even if we remove the Library,
    the books are still exist
"""
class Book():

    def __init__(self, name, author, Publish): # Its Better to write Publish in lowercase too (publish)
        self.name = name
        self.author = author
        if isinstance(Publish, int):
            self.publish = Publish
        else:
            raise TypeError("*** Publish year must be Integer ***")
        
    def info(self):
        return f"Name: {self.name}\nAuthor: {self.author}\nPublish year: {self.publish}"


class Library():
    
    def __init__(self, lib_name):
        self.lib_name = lib_name  
        self.books = []

    def add_book(self, book_obj):
        if isinstance(book_obj, Book):
            self.books.append(book_obj)
        else:
            raise TypeError("Only Book objects allowed")

    def show_books(self):
        if not self.books:
            return f"there is no books at all"
        else: 
            result = f"\n--- Welcome to {self.lib_name} Library ---\n    ---these are our books---\n\n"
            for book in self.books:
                result += f"✅ {book.name} by {book.author} -- ({book.publish})\n"
        return result

shab = Book("Shabhaye roshan", "Dastayofski", 1926)
Kimiagar = Book("Kimiagar", "Cooilo", 2006)
jenayat = Book("Jenayat va mokafat", "Dastayofski", 1929)
ivanilich = Book("Ivan Ilich death", "Kafka", 1951)
# Publish years are Fake

lib = Library("Unknown")
lib.add_book(Kimiagar)
lib.add_book(jenayat)
lib.add_book(ivanilich)
lib.add_book(shab)
print(lib.show_books())
```

## 6. Magic (Dunder) Methods

Implementing __str__ and __len__ to customize object behavior.

```python
### Basic magic(dunder) methods ###

class Playlist():

    def __init__(self, name):
        self.name = name
        self.song = []

    def add_song(self, song_name):
        self.song.append(song_name)

    def __str__(self):
        musics = f"\n--- This is {self.name} playlist ---\n\n"
        for music in self.song:
            musics += f"- {music}\n"
        return musics
    
    def __len__(self):
        return len(self.song)
    
    def remove(self, song_name):
        if song_name in self.song:
            self.song.remove(song_name)

yegane = Playlist("Mohsen Yeganeh")
yegane.add_song("Ghol")
yegane.add_song("Talkh")
yegane.add_song("Nadaramet")
yegane.add_song("Bastam")
yegane.remove("Ghol")

print(yegane)
print(f"there is {len(yegane)} songs in {yegane.name} playlist")
```

## 7. Superclass and Polymorphism

Using a common parent class with multiple subclasses.

```python
### Superclass ###
""" Animal is our superclass here
    a class that parenting 1 class or more
"""
class Animal():

    def __init__(self, name, race):
        self.name = name
        self.race = race

    def make_sound(self):
        raise NotImplementedError("Subclasses must implement this method")

class Dog(Animal):

    def __init__(self, name, race):
        super().__init__(name, race)    

    def make_sound(self):
        small_dogs = ["poppy", "maltiz", "chitso"]
        if self.race.lower() in small_dogs:
            return "Yap!Yap!"
        else:
            return "Woof!"

class Cat(Animal):

    def __init__(self, name, race):
        super().__init__(name, race)

    def make_sound(self):
        return "Meow!"

def animal_sound(obj):
    res = obj.make_sound()
    return res


pillow = Cat("Pillow", "Persian-cat")
rex = Dog("Rex", "Doberman")
alex = Dog("Alex", "Poppy")
print(animal_sound(pillow))
print(animal_sound(rex))
print(animal_sound(alex))
```

## 8. Abstract Classes

Abstract classes define a contract that subclasses must implement.

```python
### Abstract class ###

from abc import ABC , abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

class Circle(Shape):

    def __init__(self, r):
        self.r = r

    def area(self):
        pi = 3.14159265
        return self.r ** 2 * pi
    
c1 = Circle(5)
r1 = Rectangle(6,5)

print(f"{c1.area():.2f}")
print(r1.area())

```

## 9. Composition

A car contains an engine and depends on it for part of its functionality.

```python
### Composition ###
class Car():
    def __init__(self, weight, model):
        self.weight = weight
        self.model = model
        self.engine = Engine(6) # composition
    
    def engine_type(self):
        engine_type = (
            "Linear"
            if self.engine.cylinder_count == 4
            else "V"
            if self.engine.cylinder_count in [6, 8]
            else "Round"
        )

        return (
            f"The car engine with "
            f"{self.engine.cylinder_count} cylinders "
            f"must be a {engine_type}-Type"
        )
class Engine():
    def __init__(self, cylinder_count):
        self.cylinder_count = cylinder_count

    def engine_weight(self):
        return f"Your V-Type engine weight {self.cylinder_count * 40}\n'40 Killograms per cylinder'" # I don't know the exact Number 😒
```

بررسی تفاوت های چند موضوع مهم:
